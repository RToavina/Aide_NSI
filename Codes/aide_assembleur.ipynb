## 1️⃣ Rappels de cours essentiels

### 1.1 Rôle de l’assembleur

- L’**assembleur** est un langage de bas niveau, très proche du **langage machine**.
- Chaque instruction assembleur correspond (presque) à une instruction machine.
- Il permet de comprendre **comment un programme est réellement exécuté par le processeur**.

👉 En NSI, l’assembleur sert à :
- comprendre l’architecture d’un ordinateur,
- expliquer le lien entre Python et le matériel,
- raisonner sur l’exécution d’un programme.

---

### 1.2 Architecture simplifiée d’un processeur

Un processeur simple contient :

- **Registres** : mémoire très rapide intégrée au processeur
- **Mémoire** : stocke les données et les programmes
- **Compteur ordinal (PC)** : adresse de la prochaine instruction à exécuter

#### Cycle d’exécution

1. **Fetch** : le processeur lit l’instruction pointée par le PC
2. **Decode** : il interprète l’instruction
3. **Execute** : il l’exécute
4. Le PC est mis à jour

---

## 2️⃣ Instructions élémentaires

### 2.1 Chargement et stockage

```asm
LOAD R1, 5      ; R1 ← 5
STORE R1, 100   ; mémoire[100] ← R1
```

- `LOAD` charge une valeur dans un registre
- `STORE` enregistre le contenu d’un registre en mémoire

---

### 2.2 Opérations arithmétiques

```asm
LOAD R1, 7
LOAD R2, 3
ADD R1, R2      ; R1 ← R1 + R2
SUB R1, R2      ; R1 ← R1 - R2
```

Les calculs se font **uniquement avec les registres**.

---

### 2.3 Comparaison et branchements conditionnels

En assembleur vu en **Première NSI**, on utilise des **instructions de branchement conditionnel** plutôt que des sauts généraux.

```asm
CMP R1, R2      ; compare R1 et R2
B label         ; branchement inconditionnel
BE label        ; branchement si égalité
BEQ label       ; branchement si égalité (équivalent selon les syntaxes)
BNE label       ; branchement si différent
```

👉 Les branchements permettent de modifier le déroulement du programme **en fonction d’un test**.

---

## 3️⃣ Registres et mémoire

| Élément | Rôle |
|-------|------|
| Registre | Très rapide, temporaire |
| Mémoire | Plus lente, stockage |

Exemple :

```asm
LOAD R1, 10
STORE R1, 200
```

👉 Le registre est modifié immédiatement, la mémoire seulement avec `STORE`.

---

## 4️⃣ Exécution pas à pas (méthode DS)

Programme :

```asm
LOAD R1, 5
LOAD R2, 2
ADD R1, R2
STORE R1, 100
```

| Instruction | R1 | R2 | mémoire[100] |
|------------|----|----|---------------|
| LOAD R1,5  | 5  | –  | – |
| LOAD R2,2  | 5  | 2  | – |
| ADD R1,R2  | 7  | 2  | – |
| STORE R1,100 | 7 | 2 | 7 |

👉 **Classique en DS commun**.

---

## 5️⃣ Structures de contrôle

### 5.1 Condition (if)

Python :
```python
if a == b:
    x = 1
```

Assembleur :
```asm
CMP R1, R2
BNE fin
LOAD R3, 1
fin:
```

---

### 5.2 Boucle (while)

Python :
```python
while n > 0:
    n = n - 1
```

Assembleur :
```asm
debut:
CMP R1, 0
BE fin
SUB R1, 1
B debut
fin:
```

---

## 6️⃣ Traduction Python → Assembleur

Python :
```python
x = 3
y = 4
z = x + y
```

Assembleur :
```asm
LOAD R1, 3
LOAD R2, 4
ADD R1, R2
STORE R1, 100
```

👉 Toujours : **variables → registres → mémoire**.

### Énoncé

```asm
LOAD R1, 0
LOAD R2, 3
boucle:
CMP R1, R2
BE fin
ADD R1, 1
B boucle
fin:
STORE R1, 200
```

2. Compléter le tableau d’exécution
3. Expliquer le rôle du programme

### Correction

- Le programme incrémente R1 jusqu’à atteindre R2
